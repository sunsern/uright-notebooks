```python
import pickle
from uright.classifier import ClassifierDTW

filename = "new_uright_simulation_results_ipad-53.p"
data = pickle.load(open(filename, "rb"))

from IPython.core.display import HTML
HTML("<h1>User ID: %s:%s</h1>"%(data['userid'], data['username']))
```

Classification accuracy
-----------------------

```python
import numpy as np

def smooth(x, N):
    return np.convolve(x, np.ones((N,))/float(N), 'same')

saved_accu = data['accu']
num_groups = data['num_groups']
prototype_weights = data['prototype_weights']

x = np.arange(num_groups-1) + 1
figure(figsize(15,20))

subplot(211)
for proto_weight in prototype_weights:
    accuracy = np.zeros(num_groups-1)
    for i in range(num_groups-1):
        accuracy[i] = saved_accu[(proto_weight,i+1)]
    plot(x, accuracy, 'o-', label="proto_weight=%d"%proto_weight)
xticks(x)
xlabel('Session')
ylabel('Accuracy')
legend(loc=4)
title('Accuracy at each session')

if num_groups > 5:
    subplot(212)
    for proto_weight in prototype_weights:
        accuracy = np.zeros(num_groups-1)
        for i in range(num_groups-1):
            accuracy[i] = saved_accu[(proto_weight,i+1)]
        plot(x[:-4], smooth(accuracy,5)[2:-2], 'o-', label="proto_weight=%d"%proto_weight)
    xticks(x)
    xlabel('Session')
    ylabel('Average accuracy')
    legend(loc=4)
    title('average accuracy (window_size=5)')

pass
```

Prototypes
----------

```python
skip = 3
if num_groups < 10:
    skip = 1
elif num_groups < 20:
    skip = 2

dtw_list = []
for i in range(1,num_groups,skip):
    try:
        dtw_list.append(data['dtw'][(3,i)])
    except KeyError:
        break

def show_prototype(label):
    from uright.prototype import PrototypeDTW
    from uright.visualization import draw_ink

    prot_list = []
    num_prot = []
    for each_dtw in dtw_list:
        filtered_prot = []
        for p in each_dtw.trained_prototypes:
            if p.label == label:
                filtered_prot.append(p)
        prot_list.append(filtered_prot)
        num_prot.append(len(filtered_prot))

    n_row = max(num_prot)
    n_col = len(num_prot)
    row_height = 3
    col_width = 3
    fig = figure(figsize=(col_width * n_col, row_height * n_row))
    for c_idx in range(n_col):
        prots = prot_list[c_idx]
        prots = sorted(prots, key=lambda x:-x.total_weight) 
        for r_idx in range(len(prots)):
            ax = subplot(n_row, n_col, n_col*r_idx + c_idx + 1)
            draw_ink(ax, prots[r_idx].model, show_order=True)
            ax.set_xlim(-2,2)
            ax.set_ylim(2.5,-2.5)
            ax.set_xticks([])
            ax.set_yticks([])
            ax.set_title("Iter %d [%d]"%(c_idx*skip+1, prots[r_idx].total_weight))
```
