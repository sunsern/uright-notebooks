```python
users = ['ipad-1','ipad-6','ipad-9','ipad-29','ipad-32','ipad-35','ipad-40','ipad-43','ipad-50','ipad-52','ipad-53','ipad-56','iphone-1-1','iphone-1-32',
         'iphone-1-35','iphone-1-62','iphone-1-63','iphone-1-65','iphone-1-66','iphone-1-67']
```

```python
def smooth(x, N):
    return np.convolve(x, np.ones((N,))/float(N), 'same')

def show_graph(filename):
    import pickle
    import numpy as np
    from uright.classifier import ClassifierDTW

    data = pickle.load(open(filename, "rb"))

    saved_accu = data['accu']
    num_groups = data['num_groups']
    prototype_weights = data['prototype_weights']
    username = data['username']

    x = np.arange(num_groups-1) + 1
    figure(figsize(30,10))
    
    subplot(121)
    for proto_weight in prototype_weights:
        accuracy = np.zeros(num_groups-1)
        for i in range(num_groups-1):
            accuracy[i] = saved_accu[(proto_weight,i+1)]
        plot(x, accuracy, 'o-', label="proto_weight=%d"%proto_weight)
    xticks(x)
    xlabel('Session')
    ylabel('Accuracy')
    legend(loc=4)
    title('user = %s:%s'%(user_id,username))
    
    if num_groups > 5:
        subplot(122)
        for proto_weight in prototype_weights:
            accuracy = np.zeros(num_groups-1)
            for i in range(num_groups-1):
                accuracy[i] = saved_accu[(proto_weight,i+1)]
            plot(x[:-4], smooth(accuracy,5)[2:-2], 'o-', label="proto_weight=%d"%proto_weight)
        xticks(x)
        xlabel('Session')
        ylabel('Average accuracy')
        legend(loc=4)
        title('average accuracy (window_size=5)')
```

```python
for user_id in users:
    filename = "new_uright_simulation_results_%s.p"%(user_id)
    show_graph(filename)
```
