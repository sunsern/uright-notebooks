```python
users = {'user_1':'sunsern','user_6':'yoav','user_9':'jirapon','user_29':'Handghost','user_32':'techdude','user_35':'Butter'}
```

```python
def show_graph(filename, username):
    import pickle
    import numpy as np
    from uright.classifier import ClassifierDTW

    (_,accu,_,_) = pickle.load(open(filename, "rb"))

    target_weights = [0, 1, 5, 1000000]

    num_reps = 0
    for i in range(100):
        if (0,i,0) in accu: num_reps = i+1
        else: break
        
    num_groups = 0
    for i in range(100):
        if (0,0,i) in accu: num_groups = i+1
        else: break

    x = np.arange(num_groups)
    figure(figsize(15,10))
    for target_weight in target_weights:
        accu_results = np.zeros((num_reps, num_groups))
        for j in range(num_reps):
            for i in range(num_groups):
                accu_results[j,i] = accu[(target_weight,j,i)]
        mean_accuracy = np.mean(accu_results,axis=0)
        std_accuracy = np.std(accu_results,axis=0)
        errorbar(x, mean_accuracy, yerr=std_accuracy, fmt='o-', label="target_weight=%d"%target_weight)
    xticks(x)
    xlabel('Iteration')
    ylabel('Accuracy')
    title("Username: %s"%username)
    legend(loc=2) 
```

```python
for user_id, username in users.iteritems():
    filename = "uright_simulation_results_%s.p"%(user_id)
    show_graph(filename, username)
```
