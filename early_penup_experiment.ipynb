Early penup experiment
======================
During writing, when the likelihood of the target label is higher than some threshold, the user is notified to stop writng and lift the pen. In this experiment, we compare the performance of early-penup enabled system with the regular system.

```python
from uright.inkutils import (json2array,normalize_ink,filter_bad_ink)
import pickle
import random

filename = "data/candidate_prototypes_1368819209.p"
candidate_proto = pickle.load(open(filename,"rb"))

num_test_examples = 30

clustered_data = {}
label_ink_pairs = []
for label in candidate_proto.keys():
    clustered_data[label] = []
    label_data = []
    for _,examples in candidate_proto[label]:
        # process all ink associated with the cluster
        data = [np.nan_to_num(normalize_ink(json2array(ink)))
                for ink in filter_bad_ink(examples)]
        weights = [1] * len(data)
        clustered_data[label].append(zip(data,weights))
        label_data += data
            
    if len(label_data) > num_test_examples:
        sample_data = random.sample(label_data, num_test_examples)
    else:
        sample_data = label_data
    label_ink_pairs += [(label,ink) for ink in sample_data]
```

```python
from uright.classifier import ClassifierDTW
from uright._classifier_beam_early import ClassifierBeamEarly
```

```python
beam = ClassifierBeamEarly()
beam.train(clustered_data)
```

```python
beam.beam.alpha = 0.5
beam.beam.penup_penalty = 10.0
beam.beam.beam_width = 1000
beam.beam.skips = 3
```

```python
no_early_stop = 0.0
all_points = 0
for (label, ink) in label_ink_pairs:
    (stop_length, prob) = beam.early_stop(ink, label, threshold=-1)
    no_early_stop += max(log2(26)+log2(max(prob,1e-6)),0)
    all_points += ink.shape[0]
```

```python
early_stop = 0.0
saving = 0
for (label, ink) in label_ink_pairs:
    (stop_length, prob) = beam.early_stop(ink, label, threshold=1)
    early_stop += max(log2(26)+log2(max(prob,1e-6)),0)
    saving += (ink.shape[0]-1-stop_length)
```

The total time of writing (excluding reaction time) is estimated by assuming the sample points were sampled at 60 hz.

```python
writing_time_no_early_stopping = all_points / 60
writing_time_no_early_stopping
```

The total time that could have been saved by stopping early is

```python
time_saved_by_early_stopping = saving / 60
time_saved_by_early_stopping
```

Using the total log-loss based scores, the BPS of the system with early stopping is

```python
bps_with_early_stopping = early_stop / (writing_time_no_early_stopping - time_saved_by_early_stopping)
bps_with_early_stopping
```

The BPS of the system without early stopping is

```python
bps_no_early_stopping = no_early_stop / writing_time_no_early_stopping
bps_no_early_stopping
```
