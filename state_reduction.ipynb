```python
from IPython.parallel import Client
import pickle

from uright.inkutils import (json2array,normalize_ink,filter_bad_ink)

rc = Client()
dview = rc[:]

filename = "candidate_prototypes_1368819209.p"
candidate_proto = pickle.load(open(filename,"rb"))

clustered_data = {}
label_ink_pairs = []
for label in ['u']: 
    clustered_data[label] = []
    _,examples = candidate_proto[label][0]
    data = [np.nan_to_num(normalize_ink(json2array(ink)))
            for ink in filter_bad_ink(examples)]
    weights = [1] * len(data)
    clustered_data[label].append(zip(data,weights))
    label_ink_pairs += [(label,ink) for ink in data]
```

```python
from uright.classifier import ClassifierDTW

cDTW = ClassifierDTW(alpha=0.5)
cDTW.train(clustered_data)

reducedCDTW = ClassifierDTW(alpha=0.5)
reducedCDTW.train(clustered_data, state_reduction=True)
```

```python
from uright._state_reduction import _compute_fraction

labels, test_ink = zip(*label_ink_pairs)

ink0 = cDTW.trained_prototypes[0].model
score0 = _compute_fraction(ink0, test_ink)

ink1 = reducedCDTW.trained_prototypes[0].model
score1 = _compute_fraction(ink1, test_ink)
```

```python
%load_ext autoreload
%autoreload 2

from matplotlib.lines import Line2D
from matplotlib.patches import Rectangle
from uright.visualization import draw_ink

def subplot_draw(ax, ink, scores, title):
    node_colors = []
    for i in range(ink.shape[0]):
        if (scores[i] > 1.0):
            node_colors.append('#CC0000')
        elif (scores[i] > 0.75):
            node_colors.append('#FFB266')
        elif (scores[i] > 0.50):
            node_colors.append('#FFFF66')
        else:
            node_colors.append('#FFFFFF')
 
    draw_ink(ax, ink, node_colors=node_colors,
             penup_color='#0000FF',
             pendown_color='#00FF00')
    ax.set_xlim(-4,4)
    ax.set_ylim(2.5,-2.5)
    ax.set_title(title)
    
def draw_state_reduction(ink, scores, titles):
    fig = figure(figsize=(15,8))
    ax1 = fig.add_subplot(121)
    ax2 = fig.add_subplot(122)
    subplot_draw(ax1,ink[0],scores[0],titles[0])
    subplot_draw(ax2,ink[1],scores[1],titles[1])
    
    # legend
    p1 = Rectangle((0,0),1,1,alpha=0.5,fc="#CC0000")
    p2 = Rectangle((0,0),1,1,alpha=0.5,fc="#FFB266")
    p3 = Rectangle((0,0),1,1,alpha=0.5,fc="#FFFF66")
    p4 = Rectangle((0,0),1,1,alpha=0.5,fc="#FFFFFF")
    c1 = Line2D(range(1), range(1),
                color="white", marker='o', mec="#00FF00",
                mew=3, ms=10, alpha=0.5)
    c2 = Line2D(range(1), range(1),
                color="white", marker='o', mec="#0000FF",
                mew=3, ms=10, alpha=0.5)
    l1 = ax2.legend([p1,p2,p3,p4],[' > 100%','75% - 100%',
                                       '50% - 75%','0% - 50%'],
                    bbox_to_anchor=(1.1, 1), loc=2, borderaxespad=0)
    
    l2 = ax1.legend([c1,c2],['Pen-down','Pen-up'],
                    bbox_to_anchor=(-0.1, 1),
                    loc=1, borderaxespad=0)
    
```

```python
draw_state_reduction((ink0,ink1),(score0,score1),('before','after'))
```
