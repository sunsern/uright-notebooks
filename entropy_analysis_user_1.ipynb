```python
# User id 
target_user = 'user_1'
# Maximum number of clusters per label
maxclust = 2

# fix random seed
import random
random.seed(12345)
```

```python
import pickle

from uright.inkutils import (json2array,
    normalize_ink, filter_bad_ink, user_normalized_ink)
from uright.classifier import ClassifierDTW

user_raw_ink = pickle.load(open("rawink_1_1371107932.p","rb"))
user_ink_data = user_normalized_ink(user_raw_ink)

user_ink = []
labels = []
for label in user_raw_ink[target_user]:
    user_ink += user_ink_data[target_user][label]
    labels += [label] * len(user_ink_data[target_user][label]) 
    
only_target = {target_user:user_ink_data[target_user]}
```

```python
from uright.clustering import ClusterKMeans
from uright.classifier import ClassifierBeam
 
clusterer = ClusterKMeans(only_target, 
                          algorithm='dtw', 
                          maxclust=maxclust)
clustered_data = clusterer.clustered_data()
beam = ClassifierBeam(beam_width=500)
beam.train(clustered_data, center_type='centroid')
```

```python
from uright.entropy import compute_posterior, compute_entropy

ink_data = {}
for label,ink in zip(labels, user_ink):
    ink_data.setdefault(label,[]).append(ink)

post = compute_posterior(beam, ink_data)
x = compute_entropy(post)
```

```python
import numpy as np
import random
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset

figure(figsize(15,15))

max_ent_length = np.max([ent.shape[0] for ent in x.values()])

subplot(211)
for label, ent in x.iteritems():
    new_ent = np.ones(max_ent_length) * ent[-1]
    new_ent[:ent.shape[0]] = ent
    plot(ent, label=label)
    annotate(label, xy=(ent.shape[0]+random.random() - 0.5,new_ent[-1]),fontsize=20)
xlabel('time')
ylabel('H (bits)')
    
subplot(212)
for label, ent in x.iteritems():
    new_ent = np.ones(max_ent_length) * ent[-1]
    new_ent[:ent.shape[0]] = ent
    plot(new_ent, label=label)
    annotate(label, xy=(ent.shape[0],new_ent[-1]),fontsize=30)
xlabel('time')
ylabel('H (bits)')
xlim(9,40)
ylim(0,2)
```
