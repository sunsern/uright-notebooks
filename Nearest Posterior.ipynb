Q: Why the negative evidence is more reliable than the positive?

A: It's easier model the class conditional density $f$ where $f$ is 0. 


* In ova setting, if each the estimated class probability density has a probability of errors, relying on just the maximum is not a good idea.

Sufficient Conditions
==========
1. The estimated class conditional density $f$ is more reliable when $f(x)$ is small.
2. The average posterior conditioned on example from each class has small variation.

Nearest Posterior
=================

Assume a metric space $(X,d)$. $M$ classes.

Input: $(x_1, i_1),(x_2, i_2), ..., (x_n, i_n)$ and $\hat{x}$

1. Estimate $\hat{P}(I|x_j)$ for all $1 \le j \le n$ using kNN
2. Esimate $\hat{P}(I|\hat{x})$ using kNN
3. Find $j^*$ such that $D_{KL}(\hat{P}(I|\hat{x}) || \hat{P}(I|x_j)$ is minimized.
4. Predict $i_{j^*}$ 


When n $\rightarrow \infty$
===========================

* $\hat{P}(I|x) \rightarrow P(I|x)$ -- proven in kNN.
* $x_{j^*} \rightarrow \hat{x}$ -- by KL divergence property.
* $R^* \le P_e \le R^*(2 - \frac{M}{M-1}R^*)$ -- same as 1-NN case.

Advantages
==========
1. The structure of $P(I|x)$ can be exploited and incorporated into the algorithm. For example, if $P(I|x)$ for $x$ of class $i$ are concentrated and well-modelled by their mean $\tilde{P}_i$, the algorithm can be simplified to finding $i^*$ such that $D_{KL}(\hat{P}(I|\hat{x}) || \tilde{P}_i)$ is minimized.
2. There are classication algorithms that can give an estimate $\hat{P}(I|x)$ such as $GMM$, $RF$. The algorithm can be modified to finding $j^*$ such that $D_{\alpha}(\hat{P}(I|\hat{x}) || \hat{P}(I|x_j)$ is minimized over all $\alpha$.

Disadvantages
=============
1. Estimating $P(I|x)$ may not be easy and requires more examples.

Questions
=========
1. Comparing convegence rate to kNN.
2. Can we bound kNN error rate in term of Chernoff Information? Can we get a better bound of this algorithm in term of Chernoff Information?


* Asymptotic expansion of the k nearest neighbor risk

```python
def kl_divergence(p,q):
    return np.sum(p * np.maximum(np.log(p/q), 1e-99))
    
def alpha_divergence(exp, obs, alpha):
    if alpha==1:
        return kl_divergence(obs, exp)
    p = obs
    q = exp
    return (np.sum((p**alpha)*(q**(1-alpha))) - 1.0)/(alpha*(alpha-1))

def alpha_classifier(obs, P, alpha=1):
    return np.argmin([alpha_divergence(P[i,:], obs, alpha) for i in range(P.shape[0])])

def max_classifier(obs):
    return obs.argmax()
```

```python
from sklearn import cross_validation
from sklearn.neighbors import KNeighborsClassifier

def h_alpha(p, alpha=1):
    EPS = 1e-10
    if alpha==1:
        return -np.sum(p * np.log2(p + EPS))
    else:
        return np.log2(np.sum(p ** alpha) + EPS)/(1-alpha)

def prototypical_P(i, nx, Y_train, M, show_plot=False):
    temp = []
    for x in nx:
        temp.append(make_distribution(Y_train[x], M))
    plot_data = np.vstack(temp)
    m = np.mean(plot_data, axis=0)
    m[m.argmax()] = 1.0
    if show_plot:
        figure()
        subplot(111)
        boxplot(plot_data, bootstrap=1000, positions=np.arange(M))
        plot(np.arange(M), m, 'sk', label='mean')
        ylim((-0.1,1.1))
        m_minus = np.delete(m,i) + 1e-4
        m_minus = m_minus / m_minus.sum()
        title('label = %d, H = %0.2f (max = %0.2f)'%(i, h_alpha(m_minus, 1), np.log2(m_minus.shape[0])))
    return m

def eval_P(P, alpha=1, M=26):
    ad = 0
    for i in range(M):
        for j in range(M):
            if i != j:
                ad += alpha_divergence(P[i,:], P[j,:], alpha=alpha)
    return ad / (M*(M-1))

def make_distribution(Sx, M):
    dist,_  = np.histogram(Sx, bins=M, range=(0,M),density=True)
    return dist

def compute_errors_fixed_P(X_test, Y_test, Y_train, P, nn_all, kn, M, alpha):         
    correct_ll = 0
    correct_ah = 0
    for i,x in enumerate(X_test):
        nx = nn_all.kneighbors(x, n_neighbors=kn, return_distance=False)
        dx = make_distribution(Y_train[nx], M)
        ll = max_classifier(dx)
        ah = alpha_classifier(dx, P, alpha)
        if ll == Y_test[i]:
            correct_ll += 1
        if ah == Y_test[i]:
            correct_ah += 1
            
    return 1.0 - np.array([float(correct_ll)/len(X_test), float(correct_ah)/len(X_test)])

def run_experiment_fixed_P(kfold, X, Y, alpha, bias=0.01):
    test_errors = []
    skf = cross_validation.StratifiedKFold(Y, n_folds=kfold)
    M = np.max(Y).astype(int) + 1
    already_plot = False
    for train_index, test_index in skf:
        
        X_train = X[train_index]
        Y_train = Y[train_index]
        
        X_test = X[test_index]
        Y_test = Y[test_index]
            
        # train overall NN models
        nn_all = KNeighborsClassifier(warn_on_equidistant=False)
        nn_all.fit(X_train, Y_train)
        
            
        # compute test errors using different k
        test_err = []
        all_kn = [1,3,5,7,9,11,21]
        for kn in all_kn:
            
            if kn == 7:
                showPlot = True
            else:
                showPlot = False
                
            P = np.zeros((M,M))
            for j in range(M):
                idx = (Y_train == j)
                nx = nn_all.kneighbors(X_train[idx], n_neighbors=kn, return_distance=False)
                P[j,:] = prototypical_P(j, nx, Y_train, M, show_plot=(showPlot and not already_plot)) + bias
                #P[j,:] = make_distribution(Y_train[nx.reshape(-1)], M) + bias
                #P[j,:] = P[j,:] / P[j,:].sum()
                #P[j,j] = 1.0
            #print P.round(2)
            #print percentile, eval_P(P, M=M)
            
            if showPlot:
                already_plot = True
                
            # start testing     
            test_err.append(compute_errors_fixed_P(X_test, Y_test, Y_train, P, nn_all, kn, M, alpha))
            #raise
            
        print test_err
        test_errors.append(test_err)
        
    test_error_mean = np.mean(test_errors, axis=0)
    test_error_std = np.std(test_errors, axis=0)
    return (test_error_mean, test_error_std, all_kn) 
```

```python
def plot_results(ts_m,ts_sd,x):
    figure(figsize=(10,8))
    errorbar(x, ts_m[:,0], yerr=ts_sd[:,0], fmt='sb-', label='max', lw=2)
    errorbar(x, ts_m[:,1], yerr=ts_sd[:,1], fmt='or-', label='alpha-d', lw=2)
    legend()
    xlabel('size of neighborhood')
    ylabel('error rate')
    title("Test error")
    grid()
```

```python
def scale_image(x, scale=0.5):
    from scipy.misc import imresize
    n = x.shape[0]
    sx = x.reshape((sqrt(n),sqrt(n)))[2:-2,2:-2]
    small = imresize(sx, scale)
    return small.reshape(-1)

def transform_data_pca(X):
    X = np.array([scale_image(x, 0.5) for x in X])
    from sklearn.decomposition import PCA
    pca = PCA(n_components=100)
    Tx = pca.fit_transform(X)
    #print(pca.explained_variance_ratio_.cumsum()) 
    return Tx
```

```python
(ts_m,ts_sd,x) = run_experiment_fixed_P(5, transformed_data, mnist.target, alpha=1, bias=0.01)
plot_results(ts_m,ts_sd,x)
```
